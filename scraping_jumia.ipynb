```python
import pandas as pd
import numpy as np
```

```python
product_data = pd.read_csv('jumia_products.csv')
product_data.head(10)
```

```python
product_data.info()
```

```python
# changin discount column to float
def change_disc_col(df):
    df['Discount'] = df['Discount'].str.replace('%','')
    df['Discount'] = df['Discount'].astype(float)
    return df.info()

change_disc_col(product_data)

```

```python
# filtering out prices that have a range
ranged_items = product_data[product_data['Current price'].str.contains('-') | product_data['Old price'].str.contains('-')]
# removing the rows from the original dataset
new_items_df = product_data.drop(ranged_items.index)
new_items_df
```

```python
ranged_items
```

```python
split_values = ranged_items['Current price'].str.split('-', expand=True)
ranged_items.loc[:, 'price_start_value'] = split_values[0].str.strip()
ranged_items.loc[:, 'price_end_value'] = split_values[1].str.strip()
ranged_items['price_end_value'].fillna(ranged_items['price_start_value'], inplace=True)

ranged_items
```

```python
split_values = range_items['Old_price'].str.split('-', expand=True)
range_items.loc[:, 'old_price_start_value'] = split_values[0].str.strip()
range_items.loc[:, 'old_price_end_value'] = split_values[1].str.strip()
range_items['old_price_end_value'].fillna(range_items['old_price_start_value'], inplace=True)

range_items
```

```python
range_items.drop(['Current_price', 'Old_price'], axis = 1, inplace=True)
range_items
```
